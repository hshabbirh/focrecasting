```python
b = a.groupby('TS_ID', group_keys=False)
date_range = pd.date_range('2017-01-01', '2020-01-31', freq='D')

df_list = []

for key, group in b:
    # df = group.asfreq('D')
    df = group.reindex(date_range)
    print(key)
    df['TS_ID'] = key   
    df_list.append(df)
    
df = pd.concat(df_list)
```
